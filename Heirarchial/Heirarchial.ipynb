```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
df=pd.read_csv("/content/Cancer_Data.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
x=df[['radius_mean','texture_mean']]
```

```python
x
```

```python
sns.scatterplot(x=x['radius_mean'], y=x['texture_mean'])
```

```python
plt.figure(figsize=(40,20))
shc.dendrogram(shc.linkage(x,method='ward'))
plt.xticks(fontsize=20);
```

```python
a= AgglomerativeClustering(n_clusters=5)
```

```python
cluster=a.fit_predict(x)
```

```python
cluster
```

```python
sns.scatterplot(x=x[cluster==0]['radius_mean'],y=x[cluster==0]['texture_mean'])
sns.scatterplot(x=x[cluster==1]['radius_mean'],y=x[cluster==1]['texture_mean'])
sns.scatterplot(x=x[cluster==2]['radius_mean'],y=x[cluster==2]['texture_mean'])
sns.scatterplot(x=x[cluster==3]['radius_mean'],y=x[cluster==3]['texture_mean'])
sns.scatterplot(x=x[cluster==4]['radius_mean'],y=x[cluster==4]['texture_mean'])
```

```python
print(silhouette_score(x,a.labels_))
```
